```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sample_data/spotify-2023_fixed.csv")
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
Streams_list = df['streams'].tolist()
Streams_list
```

```python
for i in range(len(Streams_list)):
  if 'Danceability' in Streams_list[i]:
    Streams_list[i] = '000000000'
Streams_list
```

```python
df['streams'] = Streams_list
```

```python
df['streams'] = df['streams'].astype(int)
```

```python
in_deezer_list = df['in_deezer_playlists'].tolist()
```

```python
for i in range(len(in_deezer_list)):
  if ',' in in_deezer_list[i]:
    in_deezer_list[i] = in_deezer_list[i].replace(',','')
in_deezer_list
```

```python
df['in_deezer_playlists'] = in_deezer_list
```

```python
df['in_deezer_playlists'] = df['in_deezer_playlists'].astype(int)
```

```python
shazam_charts_list = df['in_shazam_charts'].tolist()
```

```python
for i in range(len(shazam_charts_list)):
  if isinstance(shazam_charts_list[i], str) and ',' in shazam_charts_list[i]:
    shazam_charts_list[i] = shazam_charts_list[i].replace(',','')
```

```python
df['in_shazam_charts'] = shazam_charts_list
```

```python
df['in_shazam_charts'] = df['in_shazam_charts']. fillna(0)
```

```python
df['in_shazam_charts'] = df['in_shazam_charts'].astype(int)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
title_dict = {'fontsize': 20, 'fontweight': 'bold', 'color': '#5f9ea0', 'family': 'monospace'}
lable_dict = {'fontsize': 15, 'color': '#dcdcdc', 'family': 'monospace'}
sns.set_theme(style="whitegrid")
%config InlineBackend.figure_format = 'svg'
```

```python
df.loc[df['released_year'] <= 1999, 'released_year_group'] = 'Старые'
df.loc[(df['released_year'] > 2000) & (df['released_year'] <= 2019), 'released_year_group'] = 'Нулевые и 10-е'
df.loc[df['released_year'] > 2020, 'released_year_group'] = 'Новые'
```

```python
df['total_playlists'] = df['in_spotify_playlists'] + df['in_deezer_playlists'] + df['in_apple_playlists']
```

```python
sns.scatterplot(data=df, x='released_year', y = 'total_playlists', hue = 'released_year_group', palette="pastel")
plt.title("Успешность песни в зависимости от года выхода", fontdict=title_dict)
plt.xlabel("Год выхода", fontdict=lable_dict)
plt.ylabel("Кол-во добавлений в плейлисты", fontdict=lable_dict)
plt.legend(title='Возраст песни')
sns.set(rc={'figure.figsize': (10, 6)})
sns.set_theme(style="whitegrid")
plt.show()
```

Из графика следует, что пользователи добавляют чаще в свои плейлисты песни, вышедшие с 1980 по 2020 годы. При этом мы видим, что после 2020 вышло больше песен, но  они менее популярны (*если считать популярностью добавление в плейлист на какой-либо из изучаемых платформ: spotify, deezer, apple*).

```python
sns.scatterplot(data=df, x='speechiness_%', y='total_playlists', hue='acousticness_%', palette="crest", size='danceability_%')
plt.title("Успешность песни в зависимости от кол-ва слов", fontdict=title_dict)
plt.xlabel("Кол-во слов, %", fontdict=lable_dict)
plt.ylabel("В плейлистах, кол-во", fontdict=lable_dict)
plt.show()
```

Первоначальная гипотеза о том, что наиболее популярными будут песни с бОльшим количеством слов (рэп) не подтвердилась. Наибольшей популярностью пользуются песни с небольшим количеством слов, достаточно акустические и мелодичные (не танцевальные). Вероятно, это характеристики поп-музыки.

```python
sns.histplot(data=df, x='speechiness_%', hue='released_year_group', palette="crest")
plt.title("Кол-во слов в песне в зависимости от года выхода", fontdict=title_dict)
plt.xlabel("Кол-во слов, %", fontdict=lable_dict)
plt.ylabel("Кол-во песен", fontdict=lable_dict)
plt.show()
```

Действительно, и среди новых песен большинство - с небольшим % слов.

```python
sns.scatterplot(data=df, x='valence_%', y='in_apple_charts', hue='released_year_group', palette="colorblind", size='bpm')
plt.title("Как влияет на успех песни позитивное настроение", fontdict=title_dict)
plt.xlabel("Насколько песня позитивна, %", fontdict=lable_dict)
plt.ylabel("Рейтинг Apple", fontdict=lable_dict)
plt.show()
```

То, насколько позитивным можно назвать общее настроение, никак не влияет на рейтинг песни в чартах Apple Music (как и на все остальные параметры, характеризующие ее успех).

```python
df.loc[(df['released_month'] <= 2) | (df['released_month'] == 12), 'season'] = 'Зима'
df.loc[(df['released_month'] > 2) & (df['released_month'] <= 5), 'season'] = 'Весна'
df.loc[(df['released_month'] > 5) & (df['released_month'] <= 8), 'season'] = 'Лето'
df.loc[(df['released_month'] > 8) & (df['released_month'] <= 11), 'season'] = 'Осень'
```

```python
sns.scatterplot(data=df, x = 'released_year', y = 'total_playlists', hue='season', palette="colorblind")
plt.title("Успех песни от времени года", fontdict=title_dict)
plt.xlabel("Год выхода песни", fontdict=lable_dict)
plt.ylabel("Кол-во добавлений в плейлисты", fontdict=lable_dict)
plt.show()
```

Успех песни не зависит от того, в какое время года состоялся релиз.
