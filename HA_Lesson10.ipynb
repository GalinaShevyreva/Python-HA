```python
Streams_list = df['streams'].tolist()
Streams_list
```

```python
for i in range(len(Streams_list)):
  if 'Danceability' in Streams_list[i]:
    Streams_list[i] = '000000000'
Streams_list
```

```python
df['streams'] = Streams_list
```

```python
df['streams'] = df['streams'].astype(int)
```

```python
in_deezer_list = df['in_deezer_playlists'].tolist()
```

```python
for i in range(len(in_deezer_list)):
  if ',' in in_deezer_list[i]:
    in_deezer_list[i] = in_deezer_list[i].replace(',','')
in_deezer_list
```

```python
df['in_deezer_playlists'] = in_deezer_list
```

```python
df['in_deezer_playlists'] = df['in_deezer_playlists'].astype(int)
```

```python
shazam_charts_list = df['in_shazam_charts'].tolist()
```

```python
for i in range(len(shazam_charts_list)):
  if isinstance(shazam_charts_list[i], str) and ',' in shazam_charts_list[i]:
    shazam_charts_list[i] = shazam_charts_list[i].replace(',','')
```

```python
df['in_shazam_charts'] = shazam_charts_list
```

```python
df['in_shazam_charts'] = df['in_shazam_charts']. fillna(0)
```

```python
df['in_shazam_charts'] = df['in_shazam_charts'].astype(int)
```
